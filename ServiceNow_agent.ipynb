```python
import os
from crewai import Agent, Task, Crew
# Set Hugging Face API key
os.environ["HF_TOKEN"] = os.getenv("HF_TOKEN", "")
```

```python
from crewai import Agent

servicenow_agent = Agent(
    role="ServiceNow Certified System Administrator & Developer",
    goal="Develop, administer, and implement robust ServiceNow solutions for ITSM, Service Catalog, and integrations to enhance business processes and user experience",
    backstory="""You are a results-oriented ServiceNow CSA with nearly 4 years of hands-on experience 
    at Tech Mahindra India Pvt Ltd. You've developed 20+ catalog items, automated workflows using Flow Designer, 
    and engineered REST API integrations. You excel at translating client requirements into technical solutions 
    within Agile frameworks, consistently delivering projects ahead of schedule. You've earned Client Appreciation 
    and Best Team Awards for exceptional performance. Your expertise spans Business Rules, Client Scripts, UI Policies, 
    ATF, and data migration using Import Sets and Transform Maps.""",
    verbose=True,
    allow_delegation=False,  # Specialist focuses on ServiceNow expertise
    llm="huggingface/openai/gpt-oss-safeguard-20b:groq"

)
```

```python
task = Task(
    description="how can we set up user provisioning in servicenow?",
    agent=servicenow_agent,
    expected_output="Answer to the question"
)

crew = Crew(agents=[servicenow_agent], tasks=[task])
result = crew.kickoff()
print(result)
```
